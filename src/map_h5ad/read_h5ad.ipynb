```python
import scanpy as sc
import os
import pandas as pd

# Path to your data
data_dir = "G:/Secondary_analysis/Organized_by_Tissue/Liver/"
# Or check the annotated folder specifically:
# data_dir = "F:/12_30_2025_test_azimuth_annotation/annotated/"

# List all h5ad files
files = [f for f in os.listdir(data_dir) if f.endswith('.h5ad')]

report = []

for file_name in files:
    file_path = os.path.join(data_dir, file_name)

    try:
        # read_h5ad(..., backed='r') is extremely fast because it doesn't load the whole matrix
        adata = sc.read_h5ad(file_path, backed='r')

        # Check for the label in obs
        has_label = 'azimuth_label' in adata.obs.columns

        # Check for the projection in obsm
        has_umap = 'X_umap_proj' in adata.obsm.keys()

        report.append({
            "File": file_name,
            "Cells": adata.n_obs,
            "has_label": has_label,
            "has_umap": has_umap
        })
    except Exception as e:
        report.append({"File": file_name, "Error": str(e)})

# Display as a nice table
df_report = pd.DataFrame(report)
print(df_report)
```
